### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = "SELECT Age FROM master WHERE Age IS NOT NULL"
age_df = pd.read_sql_query(QUERY, conn)

# Plot the box plot
plt.figure(figsize=(10, 6))
plt.boxplot(age_df['Age'], vert=False)
plt.title('Box Plot of Age', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.grid(axis='x', alpha=0.75)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = "SELECT Age, WorkWeekHrs FROM master WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL"
scatter_data_df = pd.read_sql_query(QUERY, conn)

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(scatter_data_df['Age'], scatter_data_df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

# Plot the bubble plot
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master 
WHERE WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL AND Age IS NOT NULL
"""
bubble_data_df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 6))
plt.scatter(bubble_data_df['WorkWeekHrs'], bubble_data_df['CodeRevHrs'], s=bubble_data_df['Age'], alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs', fontsize=15)
plt.xlabel('Work Week Hours', fontsize=12)
plt.ylabel('Code Review Hours', fontsize=12)
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
WHERE DatabaseDesireNextYear IS NOT NULL
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

databases_df = pd.read_sql_query(QUERY, conn)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(databases_df['count'], labels=databases_df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year', fontsize=15)
plt.axis('equal')
plt.show()
```
